```python
runs=pd.read_csv("./datasets/sharma-kohli.csv")
runs.head()
```

```python
#Line Plot
plt.plot(runs['index'],runs['V Kohli'],label="Virat Kolhi",marker='D')
plt.plot(runs['index'],runs['RG Sharma'],label="Rohit Sharma",marker='D')

plt.title("Runs scored by Virat and Rohit")
plt.xlabel("Years")
plt.ylabel("Runs")

plt.legend()
plt.grid()
plt.show()
```

Rohit Sharma has consistently scored over 300 runs in ipl seasons, which shows that he is a consistent player.
Virat Kolhi did not have a good first season in 2008 scoring below 200 runs. However, the next year he showed significant progress and has also been scoring runs more than 300 after 2010. The year 2016 marks as a milestone for Virat Kolhi as he scores approximately 1000 runs that is still a world record. 
Overall, Virat Kolhi has generally performed better than Rohit Sharma over the years, playing some phenomenal innings showcasing his ability.

```python
batsman=pd.read_csv("./datasets/batter.csv")
batsman=batsman.head(50)
```

```python
#Scatter Plot of top 50 ipl batsman
plt.scatter(batsman['avg'],batsman['strike_rate'])

plt.title("Strike rate and average runs of top 50 ipl batsman")
plt.xlabel("Average Runs")
plt.ylabel("Strike Rate")


plt.grid()
```

```python
record=pd.read_csv("./datasets/batsman_season_record.csv")
```

```python
record.head()
```

```python
#Single Barplot:
plt.bar(record['batsman'],record['2015'],color='red')
plt.title("Runs scored by 5 batsman in season 2015 of IPL")
plt.xlabel("Players")
plt.ylabel("Runs")

plt.show()
```

```python
#Multiple Barplot:
import numpy as np
plt.bar(np.arange(record.shape[0])-0.2,record['2015'],width=0.2)
plt.bar(np.arange(record.shape[0]),record['2016'],width=0.2)
plt.bar(np.arange(record.shape[0])+0.2,record['2017'],width=0.2)

plt.title("Runs of top 5 batsman in season 2015,2016,2017")
plt.xlabel("Players")
plt.ylabel("Runs")

plt.xticks(np.arange(record.shape[0]),record['batsman'])

plt.show()
```

David Warner is the most consistent batsman in all three seasons scoring more than 400 runs in all three of them and was the highest eun scorer in 2 of them.
However, no one has matched the runs scored ny Virat Kolhi in 2016 season which shows the class of him.

```python
#Stacked Barplot:
plt.bar(record['batsman'],record['2015'])
plt.bar(record['batsman'],record['2016'],bottom=record['2015'])
plt.bar(record['batsman'],record['2017'],bottom=(record['2015']+record['2016']))

plt.title("Runs of top 5 batsman in season 2015,2016,2017")
plt.xlabel("Players")
plt.ylabel("Runs")

plt.show()
```

With the help of stacked barplot we can also see the total runs of the players in all three seasons. David Warner has scored around 2000 runs,whereas, MS Dhoni has scored around 1000 runs. The reason might be because Warner bats in the opening and MSD bats around number 4 to 6.
